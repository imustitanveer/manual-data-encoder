# Load Data

```python
import pandas as pd

# Load the CSV data into a DataFrame
df = pd.read_csv('smmh.csv')

# Display the DataFrame to verify it's loaded correctly
print(df)
```

# Get Unique Values

```python
unique_values = df['3. Relationship Status'].unique()  # This gets the unique values
unique_count = df['3. Relationship Status'].nunique()  # This counts the number of unique values

print(f"Number of distinct values in the column: {unique_count}")
print("Unique values in the column:", unique_values)
```

```python
unique_values = df['4. Occupation Status'].unique()
unique_count = df['4. Occupation Status'].nunique()

print(f"Number of distinct values in the column: {unique_count}")
print("Unique values in the column:", unique_values)
```

```python
# Replace 'column_name' with the actual name of the column you are interested in
unique_values = df['8. What is the average time you spend on social media every day?'].unique()
unique_count = df['8. What is the average time you spend on social media every day?'].nunique()  
print(f"Number of distinct values in the column: {unique_count}")
print("Unique values in the column:", unique_values)
```

```python
# Replace 'column_name' with the actual name of the column you are interested in
unique_values = df['2. Gender'].unique()
unique_count = df['2. Gender'].nunique()  
print(f"Number of distinct values in the column: {unique_count}")
print("Unique values in the column:", unique_values)
```

# Code Values

```python
# Replace 'Male' with 0 and 'Female' with 1 and so on, in the 'Gender' column
df['2. Gender'] = df['2. Gender'].replace({'Male': 0, 'Female': 1, 'Nonbinary': 2, 'Non-binary': 2, 'NB': 2, 'Non binary': 2, 'Trans': 3, 'unsure': 4})
```

Male = 0,
Female = 1,
Non-Binary = 2,
Trans = 3,
Unsure = 4

```python
df['3. Relationship Status'] = df['3. Relationship Status'].replace({'Single': 0, 'In a relationship': 1, 'Married': 2, 'Divorced': 3})
```

Single = 0
In a relationship = 1
Married = 2
Divorced = 3

```python
df['4. Occupation Status'] = df['4. Occupation Status'].replace({'School Student': 0, 'University Student': 1, 'Salaried Worker': 2, 'Retired': 3})
```

School Student = 0,
University Student = 1,
Salaried Worker = 2,
Retired = 3

```python
df['6. Do you use social media?'] = df['6. Do you use social media?'].replace({'No': 0, 'Yes': 1})
```

Yes = 1, No = 0

```python
df['8. What is the average time you spend on social media every day?'] = df['8. What is the average time you spend on social media every day?'].replace({'Less than an Hour': 0,
                                                                                                                                                         'Between 1 and 2 hours': 1,
                                                                                                                                                         'Between 2 and 3 hours': 2,
                                                                                                                                                         'Between 3 and 4 hours': 3,
                                                                                                                                                         'Between 4 and 5 hours': 4,
                                                                                                                                                         'More than 5 hours': 5})
```

0 = Less than an Hour,
1 = Between 1 and 2 hours,
2 = Between 2 and 3 hours,
3 = Between 3 and 4 hours,
4 = Betweem 4 and 5 hours,
5 = More than 5 hours
